# Task 1
## #1

#### We have to calculate ROI for each chanel. For this reason let's calculate amount of money invested in each chanel. Then calculate amount of money customers, spent on orders on the nth day for each chanel. And finaly calculate ROI: 
#### ROI = (sum(Chanel Amount) - sum(Chanel cost))/ sum(Chanel cost)  for each chanel.

Let's calculate chanels' campaigns cost
<dr>
If campaign is not in Costs then its cost = 0 . 
<dr>
For that reason we fill all NaNs with zeros

```python
cols = ['chanel', 'Costs']
chanel_costs = pd.merge(Relations, Costs, how='left',left_on=['name', 'id_partner'], right_on=['campaign', 'id_partner'])
chanel_costs.Costs = chanel_costs.Costs.fillna(0)
chanel_costs = chanel_costs[cols].groupby('chanel', as_index=False).sum()
chanel_costs.head()
```

```python
# Add chanel column to users table
users_chanel = pd.merge(Users, Relations, how = 'outer', on=['id_partner', 'name'])
users_chanel.head()
```

Add user's orders to users
<dr>
If user didn`t make any order (Amount = NaN), then its amount = 0
<dr>
If user has Reg_Date NaN, then certain chanel doesn't have user, fill Reg_date 0 - see previous cell

```python
users_orders = pd.merge(users_chanel, Orders, how='left', left_on = ['id'], right_on = ['id_user'])
del users_chanel
users_orders.Amount = users_orders.Amount.fillna(0)
users_orders['Order Date'] = users_orders['Order Date'].fillna(0)
users_orders.Reg_date = users_orders.Reg_date.fillna(0)
users_orders.head()
```

Now let's calculate Amount of money each user spent
<dr>
Amount_i - amount for i-th day for each user

```python
cols = ['Reg_date', 'chanel', 'Order Date', 'Amount']
users_orders = users_orders[cols]
users_orders = users_orders.assign(Amount_0 = lambda x: x.Amount * (x.Reg_date >= x['Order Date']))
users_orders = users_orders.assign(Amount_6 = lambda x: x.Amount * (x.Reg_date + pd.DateOffset(6) >= x['Order Date']))
users_orders = users_orders.assign(Amount_14 = lambda x: x.Amount * (x.Reg_date + pd.DateOffset(14) >= x['Order Date']))
users_orders = users_orders.assign(Amount_30 = lambda x: x.Amount * (x.Reg_date + pd.DateOffset(30) >= x['Order Date']))
users_orders.head()
```

Calculate amount for each chanel:

```python
cols = ['chanel', 'Amount_0', 'Amount_6', 'Amount_14', 'Amount_30']
users_orders = users_orders[cols].groupby('chanel', as_index = False).sum()
users_orders.head()
```

New table ROI for each chanel 

```python
ROI = pd.merge(users_orders, chanel_costs, how = 'outer', on = 'chanel')
del users_orders
del chanel_costs
ROI = ROI.assign(ROI_0=lambda x: x.Amount_0/x.Costs - 1)
ROI = ROI.assign(ROI_6=lambda x: x.Amount_6/x.Costs - 1)
ROI = ROI.assign(ROI_14=lambda x: x.Amount_14/x.Costs - 1)
ROI = ROI.assign(ROI_30=lambda x: x.Amount_30/x.Costs - 1)
ROI.head()
```

If chanel's Amount and Cost = 0, then its ROI = NaN, but actualy it is 0, so fill NaN with zeros.
<dr>
Inf mean that chanel is free and its users spend money

```python
cols = ['chanel', 'ROI_0', 'ROI_6', 'ROI_14', 'ROI_30']
ROI = ROI[cols]
ROI_cols = [ 'ROI_0', 'ROI_6', 'ROI_14', 'ROI_30']
ROI = ROI[ROI_cols].fillna(0)
```

```python
ROI_per_chanel_path =  output_path + '\\ROI_per_chanel.csv'
ROI.to_csv(ROI_per_chanel_path)
```

## #2

#### Calculate average cost for all users
#### Then caluclate their first Order
#### Acording to that data calculate average cost per user

remove all previous DataFrames:

```python
if 'ROI' in globals():
    del ROI
```

Calculate number of users, that signed up on the same day and write this number for each user

```python
users_num = Users[['Reg_date', 'id_partner', 'name']].assign(num_users=lambda x: 1)
users_num = users_num.groupby(['id_partner', 'name', 'Reg_date'], as_index=False).sum()
users_cost = pd.merge(Users, users_num, how = 'left', on=['Reg_date', 'id_partner', 'name'])
del users_num
```

Add new column with cost on registration day from table Costs

```python
left_cols = ['Reg_date', 'id_partner', 'name']
right_cols = ['date_created', 'id_partner', 'campaign']
users_cost = pd.merge(users_cost, Costs, how = 'left', left_on=left_cols, right_on=right_cols)
users_cost.head()
```

Calculate average cost for user.
<dr>
If Cost = NaN, then user is free

```python
users_cost.Costs = users_cost.Costs.fillna(0)
users_cost.Costs = users_cost.Costs / users_cost.num_users
users_cost = users_cost[['id', 'Reg_date', 'id_partner', 'name', 'Costs']]
users_cost.head()
```

For each user we find order with earliest Order Date

```python
users_orders = pd.merge(users_cost, Orders, how='left', left_on = ['id'], right_on = ['id_user'])
del users_cost
print('merged')
cols = ['id', 'Reg_date', 'id_partner', 'name', 'Costs', 'Order Date']
users_orders = users_orders[cols]
users_orders = users_orders.groupby(['id', 'Reg_date', 'id_partner', 'name'], as_index=False).min()
users_orders.head()
```

Add indicators for users, acording to thier first order date

```python
users_orders = users_orders.assign(Order_14 = lambda x: 1 * (x.Reg_date + pd.DateOffset(14) >= x['Order Date']))
users_orders = users_orders.assign(Order_30 = lambda x: 1 * (x.Reg_date + pd.DateOffset(30) >= x['Order Date']))
users_orders = users_orders.assign(Order_60 = lambda x: 1 * (x.Reg_date + pd.DateOffset(60) >= x['Order Date']))
users_orders.head()
```

Add indicators of cost for users, acording to thier first order date

```python
users_orders = users_orders.assign(Cost_14 = lambda x: x.Costs * (x.Reg_date + pd.DateOffset(14) >= x['Order Date']))
users_orders = users_orders.assign(Cost_30 = lambda x: x.Costs * (x.Reg_date + pd.DateOffset(30) >= x['Order Date']))
users_orders = users_orders.assign(Cost_60 = lambda x: x.Costs * (x.Reg_date + pd.DateOffset(60) >= x['Order Date']))
users_orders.head()
```

less columns

```python
cols = ['id_partner', 'Order_14', 'Order_30', 'Order_60', 'Cost_14', 'Cost_30', 'Cost_60']
users_orders = users_orders[cols]
users_orders.head()
```

Group data by id_partner and sum of columns give us number of users, who made orders and thir costs
<dr>
Then aluclate average Cost for user.
<dr>
NaNs mean absence of orders in this chanel

```python
users_orders = users_orders.groupby('id_partner', as_index=False).sum().sort_values('Cost_14')
users_orders.Cost_14 = users_orders.Cost_14/users_orders.Order_14
users_orders.Cost_30 = users_orders.Cost_30/users_orders.Order_30
users_orders.Cost_60 = users_orders.Cost_60/users_orders.Order_60
users_orders
```

less columns and saving

```python
cols = ['id_partner', 'Cost_14', 'Cost_30', 'Cost_60']
users_orders = users_orders[cols]

cost_per_user_per_partner_path =  output_path + '\\cost_per_user_per_partner.csv'
users_orders.to_csv(cost_per_user_per_partner_path)
```

## Retention lifetime

#### Calculate the length of the period when users visited the site
#### Then claculate mean vlue per chanel

```python
iter_visits = pd.read_csv(visits_path, iterator=True, chunksize=1000000)

users_visits = Users.assign(visits=lambda x: pd.Timedelta(0, unit='d'))
users_visits = last_visit(iter_visits, users_visits, 0)
```

```python
users_visits.sort_values('visits', ascending=False).head()
```

```python
users_visits = pd.merge(users_visits, Relations, how='outer', on=['name', 'id_partner'])
users_visits = users_visits[['visits', 'chanel']]
users_visits.head()
```

```python
users_visits.visits =(users_visits.visits.fillna(0)/ np.timedelta64(1, 'D')).astype(int)
users_visits.head()
```

```python
users_visits = users_visits.groupby('chanel', as_index = False).mean()

retention_lifetime_path =  output_path + '\\retention_lifetime.csv'
users_visits.to_csv(retention_lifetimel_path)
```
