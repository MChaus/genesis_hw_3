```python
import pandas as pd
import os
import sys
```

```python
class path_error(Exception):
    pass
```

```python
project_path = os.path.dirname(os.path.realpath('__file__'))
input_path = project_path + '\\input'
processed_path = project_path + '\\processed'
output_path = project_path + '\\output'

if not os.path.exists(input_path):
        raise path_error('\n"{}"  doesn\'t exist \nPlease, check its existence with all input data'.format(input_path))
        
costs_path = input_path + '\\Costs.csv'
orders_path = input_path + '\\Orders.csv'
relations_path = input_path + '\\Relations.csv'
users_path = input_path + '\\Users.csv'
visits_path = input_path + '\\Visits.csv'

file_paths = [
    costs_path, 
    orders_path,
    relations_path,
    users_path, 
    visits_path,
]

for file_name in file_paths:
    if not os.path.exists(file_name):
        raise path_error('\n"{}"  doesn\'t exist \nplease, check its existence with all input data'.format(file_name))

print('All files were successfully loaded!')
```

```python
Cost = pd.read_csv(costs_path)
Cost.__name = 'Costs'
Orders =  pd.read_csv(orders_path)
Orders.__name = 'Orders'
Relations =  pd.read_csv(relations_path)
Relations.__name = 'Relations'
Users =  pd.read_csv(users_path)
Users.__name = 'Users'
Visits = pd.read_csv(visits_path, nrows = 1000)
Visits.__name = 'Visits'
```

```python
data_frames = [
    Cost,
    Orders,
    Relations,
    Users,
    Visits
]

for data_frame in data_frames:
    print(data_frame.__name)
    display(data_frame.head())
```

```python
order_date = 'Order Date'
df = pd.merge(Users, Orders, left_on = ['id'], right_on = ['id_user'])
df['Order Date'] = pd.to_datetime(df['Order Date']).dt.date
df['Reg_date'] = pd.to_datetime(df['Reg_date']).dt.date
```

```python
df = df[df['Reg_date'] + pd.DateOffset(6) == df['Order Date']]
```

```python
df2 = df[['id_partner', 'name', 'Amount']].groupby(['id_partner', 'name'], as_index=False).sum()
```
